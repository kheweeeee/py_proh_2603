```python
# pandas : 파이썬에서 데이터 분석과 데이터 처리를 위한 대표적인 라이브러리
# 주로 표 형태의 데이터 (엑셀, csv, 데이터 베이스 등)을 다루는 데 사용
# 데이터 과학, 머신러닝, 금융 분석, 통계 처리 등에서 매우 많이 사용됨
```

```python
import pandas as pd

# DataFrame : 여러 개의 열(column)으로 구성된 "표"형태의 데이터
#   -"열 이름" : [값들]
#   - 각 열에는 열 이름이 있음
#   - 각 행에는 지동으로 행 번호(index)가 붙는다

# 딕셔너리 형태로 DataFrame을 만듦
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
print(df)

# 열 하나씩 출력
# 결과는 DataFrame 이 아니라, "Series"라는 1차원 자료구조로 나옴
print(df["Name"])
print(df["Age"])
print(df["Sex"])

# Series : 1차원 데이터 (열 하나)
# DataFream : 여러 개의 Series가 모여서 만들어진 2차원 표 형태의 데이터
print(type(df["Age"]))
print(type(df))
```

```python
# 이름이 Age인 Series 생성
ages = pd.Series([22,35,58], name="Age")

# DataFream에서 특정 열 꺼내기
print(df["Age"])

# 기본 통계 함수
# max : 최댓값
print("최대 나이 :", df["Age"].max())

# min : 최솟값
print("최소 나이 :", df["Age"].min())

# mean : 평균
print("평균 나이 :", df["Age"].mean())

# 요약 통게 정보 보기
# describe : 숫자형 열에 대해 "요약 통계 정보"를 한 번에 보여줌
print(df.describe())

# DataFream의 전체 요약 정보
# info : DataFream의 "구조 정보"를 보여줌
print(df.info())
```

```python
# Datafream에서 특정 프레임을 선택하는 방법 2가지
# - iloc : "위치(position)"로 선택(0번째, 1번째 처럼 순서/번호로 접근)
# - loc : "라벨(label)"로 선택 "인덱스에 붙어잇는 이름으로 접근"

# iloc
# iloc은 index의 '값'이 뭐든 상관 없이, '몇 번째 줄이냐'로 가져옴
# ex) df.iloc[0] => 첫 번째 행 (0번 위치)
# 결과는 '한 행' 이므로 Series(1차원) 형태로 나옴
print(df.iloc[0])

# loc
# loc은 'index 라벨'로 접근
# 현재는 iloc과 loc의 결괴가 같음
# index를 '이름'으로 바꾸면 (ex: Name열을 index로 지정)
# log은 아름을 기준으로 접근 할 수 있음
# df.index['사람1','사람2']
# df.log['사람1']로 접근
print(df.loc[0])
```

```python
# head(): 위쪽 (처음 부분) 몇 개의 행만 보여주는 함수
# tail(): 아래쪽 (마지막 부분) 몇 개의 행만 보여주는 함수

# 상위 5개
print(df.head(2))
# head(2)는 위에서 부터 2번째 행만 출력
# 데이터가 클 때 일부분만 빠르게 확인하는 경우 자주 사용

# 하위 5개
print(df.tail(2))
# tail(2)는 마지막에서 부터 2개의 행만 출력
# 최근 데이터나 마지막 데이터 확인할 떄 자주 사용
```

```python
# 조건 필터링(Booolean Indexing)

# 나이가 30잉상인 사람만 보기
print(df[df["Age"]]>=30)
```
