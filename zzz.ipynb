```python
# 빈 클래스 만들기
class person:
    def hello(self):
        print("hello")

    def hi(self):
        self.hello()  # 클래스 내에서 다른 메서드 호출

jin = person()
jin.hi()
```

```python
class person:
    def __init__(self):
        self.hello =  "안녕하세요"

    def hi(self):
        print(self.hello)

jin = person()   # 이 때 __init__ 메서드는 이미 실행됐다
jin.hi()
```
